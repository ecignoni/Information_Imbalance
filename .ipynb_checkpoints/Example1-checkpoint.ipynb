```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from information_imbalance import information_imbalance
```

# Equivalence

```python
xx = np.linspace(0, 1, 100)
yy = xx.copy()
data = np.c_[xx, yy]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4,4))
plt.scatter(data[:,0], data[:,1], c='gray')
```

```python
print('x->xy =', information_imbalance(data[:,0], data), 'xy -> x =', information_imbalance(data, data[:,0]))
print('y->xy =', information_imbalance(data[:,1], data), 'xy -> y =', information_imbalance(data, data[:,1]))
print('y->x =', information_imbalance(data[:,1], data[:,0]), 'x -> y =', information_imbalance(data[:,0], data[:,1]))
```

# Inclusion

```python
n = 1000
xx = np.linspace(0, 1, n)
yy = 1-(1-xx)**2 - abs(np.random.normal(scale=0.3, size=n))
yy = [y if y > 0 else 0 for y in yy]
data = np.c_[xx, yy]

fig, ax = plt.subplots(1, 1, figsize=(4,4))
plt.scatter(data[:,0], data[:,1], c='gray')
```

```python
print('x->xy =', information_imbalance(data[:,0], data), 'xy -> x =', information_imbalance(data, data[:,0]))
print('y->xy =', information_imbalance(data[:,1], data), 'xy -> y =', information_imbalance(data, data[:,1]))
print('y->x =', information_imbalance(data[:,1], data[:,0]), 'x -> y =', information_imbalance(data[:,0], data[:,1]))
```

# Orthogonality

```python
n = 1000
xx = np.random.uniform(0, 1, n)
yy = np.random.uniform(0, 1, n)
data = np.c_[xx, yy]

fig, ax = plt.subplots(1, 1, figsize=(4,4))
plt.scatter(data[:,0], data[:,1], c='gray')
```

```python
print('x->xy =', information_imbalance(data[:,0], data), 'xy -> x =', information_imbalance(data, data[:,0]))
print('y->xy =', information_imbalance(data[:,1], data), 'xy -> y =', information_imbalance(data, data[:,1]))
print('y->x =', information_imbalance(data[:,1], data[:,0]), 'x -> y =', information_imbalance(data[:,0], data[:,1]))
```
